```python
df.Ca = df.Ca.astype(str)
df["Ca"] = df["Ca"].replace({"Tr": None})
df["Ca"] = df["Ca"].str.replace(',', '.').astype(float)
df.Ca.astype(float)

df.Cu = df.Cu.astype(str)
df["Cu"] = df["Cu"].replace({"Tr": None})
df["Cu"] = df["Cu"].str.replace(',', '.').astype(float)
df.Cu.astype(float)

df.Mn = df.Mn.astype(str)
df["Mn"] = df["Mn"].replace({"Tr": None})
df["Mn"] = df["Mn"].str.replace(',', '.').astype(float)
df.Mn.astype(float)
```
